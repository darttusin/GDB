```python

# importing the modules.
import xml.etree.ElementTree as ET
import re
# parsing directly.

def find_all_blocks(path: str) -> dict:
    result = {}
    tree = ET.parse('document.xml')
    current_block = None
    current_paragraph = None

    # Идем по тегам
    for el in tree.iter():
        if el.tag == '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}p':
            current_paragraph = el
            if current_block is not None:
                result[current_block]['paragraphs'].append(current_paragraph)
            print('PARAGRAPH + ', el.tag)
            continue
        # Ищем вордовские поля
        if el.tag == '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}instrText':
            text = str(el.text)
            # Если текущий текст начало блока, создаем запись в словаре, и устанавливаем текущий блок
            if text == 'block':
                print('BLOCK + ', el.tag)
                # Если текущий блок уже стоит, но найден еще один, ставим что это закрывающий тег
                if current_block is not None:
                    result[current_block]['content'].append(el)
                    print('CURRENT BLOCK IS NOT NONE, CLEARING')
                    current_block = None
                    continue
                result[el] = {'level': 0,
                              'content': [el],
                              'paragraphs': [current_paragraph]}
                print('AFTER SETTING + ', result[el])
                current_block = el
                continue
            # Записало блок и уровень в одну ячейку
            elif re.search(str(r'block\s*\[(\d{1,3})\]'), string = text):
                result[el] = {'level': int(re.findall(r'\d+', text)[0]),
                              'content': [el],
                              'paragraphs': [current_paragraph]}
                current_block = el
                continue
            # Если это не блок, скорее всего уровень. Смотрим на паттерн, записываем в поле level уровень
            elif re.search(str(r'\[\d{1,3}\]'), string = text):
                print(f'LEVEL + {el.text} + CURRENT BLOCK: {current_block}')
                if current_block is not None:
                    result[current_block]['level'] = int(re.findall(r'\d+', text)[0])
                    result[current_block]['content'].append(el)
                    continue
            # Если не блок и не уровень, то теги с информацией
            else:
                print(el.text)
                if current_block is not None:
                    result[current_block]['content'].append(el)
        # Ищем теги с текстом (реальное содержимое)
        if el.tag == '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}t':
            # Добавляем в content текущего блока тег с текстом
            if current_block is not None:
                print(el.text)
                result[current_block]['content'].append(el)
    return result

print(find_all_blocks('document.xml'))
```
